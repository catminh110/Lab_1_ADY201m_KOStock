# Lab 1 - Phần 1: Business Understanding & Analytic Approach

## 1.1. Mô tả bài toán và Mục tiêu

- **Bài toán:** Thị trường chứng khoán luôn biến động và khó dự đoán. Việc ra quyết định đầu tư đòi hỏi sự phân tích kỹ lưỡng dựa trên dữ liệu lịch sử. Bài toán đặt ra là xây dựng một quy trình khoa học dữ liệu để phân tích và dự báo giá cổ phiếu của công ty Coca-Cola (mã: KO).
- **Mục tiêu:** Sử dụng dữ liệu lịch sử để xây dựng và so sánh các mô hình dự báo, từ đó chọn ra mô hình tốt nhất để hỗ trợ việc dự báo giá trong ngắn hạn.

## 1.2. Câu hỏi Nghiên cứu & KPIs

- **Câu hỏi nghiên cứu:**
  1. Xu hướng giá và khối lượng giao dịch của KO diễn biến ra sao?
  2. Mối tương quan giữa giá đóng cửa và khối lượng giao dịch là gì?
  3. Mô hình nào (ARIMA, RandomForest,...) dự báo giá chính xác nhất?
- **KPIs:**
  - **MAE (Mean Absolute Error):** Sai số tuyệt đối trung bình.
  - **RMSE (Root Mean Squared Error).**
  - **Mục tiêu:** Đạt **MAE < 10%**.
