```python
# Tải dữ liệu đã được xử lý từ Phần 2
filename = "KO_processed_final.csv"
stock_data_sql = pd.read_csv(filename, parse_dates=True)
print("Đã tải dữ liệu thành công.")

# Chuẩn bị cho truy vấn SQL
pysqldf = lambda q: sqldf(q, globals())

# Viết và thực thi truy vấn thống kê
query_daily_trend = """
SELECT
    strftime('%Y-%m-%d', Date) AS Ngay,
    AVG(Volume) AS TrungBinhKhoiLuong,
    AVG(Close) AS TrungBinhGiaDongCua
FROM
    stock_data_sql
GROUP BY
    Ngay
ORDER BY
    Ngay;
"""
daily_trend_results = pysqldf(query_daily_trend)
print("\n--- Kết quả truy vấn thống kê hàng ngày ---")
print(daily_trend_results.head())

# Viết và thực thi truy vấn biến động giá
query_price_changes = """
SELECT
    strftime('%Y-%m-%d', Date) AS Ngay,
    (High - Low) AS BienDoGiaTrongNgay,
    (Close - Open) * 100.0 / Open AS PhanTramThayDoiGiaMoDongCua
FROM
    stock_data_sql
WHERE Open > 0
ORDER BY
    Ngay;
"""
price_changes_results = pysqldf(query_price_changes)
print("\n--- Kết quả truy vấn biến động giá ---")
print(price_changes_results.head())


# Lưu kết quả truy vấn ra file để Phần 5 sử dụng
daily_trend_results.to_csv("daily_stock_stats_KO.csv", index=False)
price_changes_results.to_csv("daily_price_changes_KO.csv", index=False)
print("\nĐã lưu kết quả các truy vấn SQL ra file CSV.")
```
