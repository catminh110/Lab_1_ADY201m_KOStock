```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

print("Các thư viện cần thiết đã được nhập.")
```

```python
# Tải dữ liệu đã được xử lý từ Phần 2
filename = "KO_processed_final.csv"
df = pd.read_csv(filename, index_col='Date', parse_dates=True)
print("Đã tải dữ liệu thành công.")

# Phân tích mô tả
print("\n--- Phân tích mô tả bằng Python ---")
print(df[['Close', 'Volume', 'Return_CC']].describe())

# Phân tích tương quan giữa Volume và Price
print("\n--- Phân tích tương quan giữa Volume và Price ---")
correlation_matrix = df[['Close', 'Volume']].corr()
print(correlation_matrix)

# Trực quan hóa ma trận tương quan
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Tương quan giữa Giá Đóng cửa và Khối lượng', fontsize=15)
plt.show()
```
