```python
!pip install yfinance
import yfinance as yf
import pandas as pd
import numpy as np

print("Các thư viện cần thiết đã được cài đặt và nhập.")
```

```python
# 2.1. Data Collection
ticker_symbol = "KO"
start_date = "2025-05-01"
end_date = "2025-06-01"

df_raw = yf.download(ticker_symbol, start=start_date, end=end_date)
print(f"Đã tải {len(df_raw)} bản ghi cho mã {ticker_symbol}.")
print("\n5 dòng dữ liệu gốc:")
print(df_raw.head())

# Flatten the multi-level column index
df_raw.columns = df_raw.columns.get_level_values(0)

# 2.2. Data Preparation
# Chuyển đổi kiểu dữ liệu, xử lý NaN
for col in ['Open', 'High', 'Low', 'Close', 'Volume']:
    df_raw[col] = pd.to_numeric(df_raw[col], errors='coerce')
df_raw.index = pd.to_datetime(df_raw.index)
df_raw.dropna(inplace=True)
if 'Adj Close' in df_raw.columns:
    df_raw = df_raw.drop(columns=['Adj Close'])

# 2.3. Feature Engineering
df_processed = df_raw.copy()
df_processed['Return_CC'] = df_processed['Close'].pct_change()
df_processed.dropna(inplace=True)

print("\nĐã hoàn tất xử lý và chuẩn bị dữ liệu.")
print(f"Số lượng bản ghi cuối cùng: {len(df_processed)}")

# 2.4. Lưu dữ liệu đã xử lý ra file CSV
output_filename = "KO_processed_final.csv"
df_processed.to_csv(output_filename)
print(f"\nDữ liệu đã xử lý được lưu vào file: {output_filename}")
```
